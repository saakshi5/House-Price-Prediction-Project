<font size="6">5.Splitting Data into x and y<font>

```python
x=df.iloc[:,:-2]
x
```

```python
y=df['Price']
```

```python
y
```

<font size="6">6.Splitting x and y into train and test<font>

```python
from sklearn.model_selection import train_test_split
xtrain,xtest,ytrain,ytest=train_test_split(x,y,test_size=0.25,random_state=1)
```

```python
xtest
```

<font size="6">7.Linear Regression Model<font>

```python
from sklearn.linear_model import LinearRegression
```

```python
#training the data
lr=LinearRegression()
z=lr.fit(xtrain,ytrain)
z
```

```python
#testing unseen data
ypred=lr.predict(xtest)
print(ypred)
```

```python
#analysing actual and predicted values
df2 = pd.DataFrame({'Actual': ytest, 'Predicted': ypred})
df2
```

```python
#Checking accuracy of model
from sklearn.metrics import r2_score,mean_squared_error
r2_score(ytest,ypred)
```

```python
#checking first 5 actual values
print(np.log(ytest[:5]))
```

```python
#checking first 5 predicted values
print(np.log(ypred[:5]))
```

```python
#Checking evaluation of model
mean_squared_error(ytest,ypred)
np.sqrt(mean_squared_error(ytest,ypred))
np.log(np.sqrt(mean_squared_error(ytest,ypred)))
```

<font size="6">8.Conclusion<font>

The R-squared (R2) score of the model is approximately 0.920, indicating that model has a relatively high R-squared score, suggesting that it performs well in explaining the variance in house prices.<br>
Actual house prices range from approximately 13.64 to 14.30 and 
Predicted house prices range from approximately 13.76 to 14.2,indicating that predicted house prices closely align with the actual prices.<br>
Finally, the root mean squared error (RMSE) of the model is approximately 11.54, indicates the average difference between the actual and predicted house prices with lower values indicating better model performance..
